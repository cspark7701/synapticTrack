```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def gaussian(x, a, x0, sigma, offset):
    return a * np.exp(-(x - x0)**2 / (2 * sigma**2)) + offset

def analyze_alison_scanner(filename, plot=True, verbose=False):
    data = np.loadtxt(filename, skiprows=1)

    x_angle = data[:, 1]  # [mrad]
    x_current = data[:, 2]  # [A]

    # Remove duplicate angle points
    unique_angles, idx = np.unique(x_angle, return_index=True)
    x_angle = x_angle[idx]
    x_current = x_current[idx]

    # Calculate RMS divergence
    weights = np.abs(x_current)
    if np.sum(weights) == 0:
        raise ValueError("All current values are zero.")
    x_center = np.sum(x_angle * weights) / np.sum(weights)
    x_rms = np.sqrt(np.sum(weights * (x_angle - x_center)**2) / np.sum(weights))

    # Try Gaussian fit, but allow failure
    threshold = 0.01 * np.max(np.abs(x_current))  # 1% of peak current
    mask = np.abs(x_current) > threshold
    x_fit = x_angle[mask]
    y_fit = np.abs(x_current[mask])

    fitted_sigma = None
    try:
        popt, _ = curve_fit(
            gaussian, x_fit, y_fit,
            p0=[np.max(y_fit), x_center, x_rms, 0],
            maxfev=20000  # large for experimental data
        )
        fitted_sigma = np.abs(popt[2])
    except RuntimeError as e:
        if verbose:
            print(f"[WARNING] Gaussian fit failed: {e}")

    if plot:
        plt.figure(figsize=(8, 5))
        plt.plot(x_angle, np.abs(x_current), 'o', label='Data')
        if fitted_sigma is not None:
            plt.plot(x_angle, gaussian(x_angle, *popt), '-', label='Gaussian Fit')
        plt.xlabel("X Angle [mrad]")
        plt.ylabel("Current [A] (abs)")
        plt.title("Alison Scanner X Angular Profile")
        plt.legend()
        plt.grid(True)
        plt.show()

    results = {
        "Angular_Center_mrad": x_center,
        "RMS_Spread_mrad": x_rms,
        "Gaussian_Sigma_mrad": fitted_sigma if fitted_sigma is not None else "Fit Failed"
    }

    return results
```
